### URL에 파라미터(Parameters) 추가하기

https://주소/주소?key2=value2&key1=value1 와 같은 URL 주소 형식을 살펴보자.

위의 방식은 쿼리 스트링(Query String)을 이용한 데이터 전달 방식으로, 클라이언트가 입력한 데이터를 서버로 전달하는 방식 중 하나이다. 

물음표 뒤에 나타나는 데이터를 쿼리 스트링이라 부르며 "key=value"의 형식을 취한다.

클라이언트가 입력하는 데이터가 바뀌면 쿼리 스트링도 바뀌게 되며, 그 결과 서버로부터 받는 응답도 달라지게 된다.

> ex) 쿼리스트링을 "key=소이연남"으로 수정하면, '소이연남'을 검색한 창이 나온다.
